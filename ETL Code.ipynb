```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
calendar_file = "calendar.csv"
calendar_df = pd.read_csv(calendar_file)
calendar_df.head()
```

```python
# convert to boolean
calendar_df["available"].replace(["t","f"], [True,False], inplace=True)
```

```python
# remove dollar sign
calendar_df["price"] = calendar_df["price"].replace({'\$': '', ',': ''}, regex=True)
# convert the price column to numeric values 
calendar_df["price"] = pd.to_numeric(calendar_df["price"])
calendar_df.head()
```

```python
listings_file = "listings.csv"
listings_df = pd.read_csv(listings_file)
listings_df.head()
```

```python
reviews_file = "reviews.csv"
reviews_df = pd.read_csv(reviews_file)
reviews_df.head()
```

```python
#merge listings and calendar
listings_calendar_df = listings_df.merge(calendar_df, left_on='id', right_on='listing_id')

#sanity check
listings_calendar_df.head()
```

```python
#choose only the necessary columns
cleaned_df = listings_calendar_df[["id", "listing_url", "name", "price_y", "number_of_reviews", "review_scores_rating", "date", "available"]]

#sanity check
cleaned_df.head()
```

```python
#drop na's
dropped_listings_df = cleaned_df.dropna(how='any', axis=0)

#sanity check
dropped_listings_df
```

```python
#show only those with rating greater than 90

#index of rows with rating greater than 90
indexNames = dropped_listings_df[dropped_listings_df['review_scores_rating'] < 90.0].index


#drop the lisitngs with rating greater than 90
dropped_listings_df.drop(indexNames , inplace=True)

#sanity check
dropped_listings_df.head()
```

```python
#drop by duplicate dates and keep the latest
dropped_listings_df.sort_values('date').drop_duplicates('id',keep='last')
```
